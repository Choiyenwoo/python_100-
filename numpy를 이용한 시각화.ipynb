### [문제]
*subplot을 이용해서 1행 2열 구조 2개씩

*4가지 색상 다르게

*4가지 스타일 다르게

*4가지 마커 서로 다르게

*범례표시

*x,y제목

```python
t = np.arange(0,12,0.5)
y1=np.sin(t)
y2=np.cos(t)
plt.grid()
plt.plot(t, y1, 'y--', marker='o')
plt.plot(t, y2, 'r--', marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Test')
plt.legend(['x','y'])
```

```python
x=np.arange(0,5,1)
x
```

```python
y1=x+1
y1
```

```python
y2=x+2
y2
```

```python
y3=x+3
y3
```

```python
y4=x+4
y4
```

```python
plt.subplot(1,2,1)
plt.plot(x, y1, 'y--', marker='.')
plt.plot(x, y2, 'b--', marker='o')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y1','y2'])

plt.subplot(1,2,2)
plt.plot(x, y3, 'g', marker='^')
plt.plot(x, y4, 'y', marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y3','y4'])
```
