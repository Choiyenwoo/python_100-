```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

### 텍스트 가져오기

```python
text = open(r'애국가(가사).txt').read()
```

```python
text
```

```python
img = r'little.png'
```

```python
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
font_path = 'C:\Windows\Fonts\malgun.ttf',
background_color = 'white',
mask=lp_mask,
max_font_size=50,
contour_width=5,
contour_color='pink').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```

```python
lp_wc.to_file(r'스폰지밥.png')
```
