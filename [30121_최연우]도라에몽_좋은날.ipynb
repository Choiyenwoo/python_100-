```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
text = open(r'좋은 날(가사).txt').read()
```

```python
text
```

```python
img = r'도라에몽.jpg'
```

```python
Image.open(img)
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
font_path = 'C:\Windows\Fonts\gulim.ttf',
background_color = 'white',
mask=lp_mask,
max_font_size=30,
contour_width=5,
contour_color='orange').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```
